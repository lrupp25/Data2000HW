```python
import pandas as pd

car_data_raw = pd.read_csv("https://cdn.c18l.org/vehicles_lab.csv")
car_data_raw.head()
```

Part 1: the features I chose were the car's age (posting_date's year minus year), odometer, condition, title_status, and transmission. I chose these, based on the fact that in my eyes, the first four absolutely are important in determining the quality and how well the car will drive. Older cars with higher odometers, usually, from my own experience, will be lower priced due to the lower reliability, and the same goes for the condition and title status. I also chose transmission because in some cases it can make a difference in price, usually from my experience making the prices higher in higher quality cars.

```python
select_data = car_data_raw.loc[:, [
    'price', 'year', 'condition', 'odometer',
    'title_status', 'posting_date', 'transmission',

]]
select_data.head()
```

Part 2:
The first, and easy choice is to remove the outliers in price, firstly removing those where the price is 0 and those where the price is higher than 25,000.

```python
data = select_data.loc[select_data['price'] > 0]
```

```python
data = data.loc[data['price'] < 25000]
```

```python
data.head()
```

Removing any year outliers as well as 2022 cars to avoid negatives being in the regression line as well as the rarities.

```python
data['year'].describe()
```

```python
data['year'].unique()
```

```python
data = data.loc[data['year'] > 1965]
data = data.loc[data['year'] < 2022]
```

```python
data['transmission'].unique()
```

Removing all rows containing nan values as to not ruin logistic regression, amount of rows available after .dropna makes imputation to not be needed.

```python
data = data.dropna()
data.describe()
```

Part 3: DataFrames I will make will be age, calculated by the posting_date minus the year column. (stolen from the code from class, respectfully), and then creating a set of dummy encoded columns for the condition.

```python
data['posting_year'] = data['posting_date'].str.extract(r'^([0-9][0-9][0-9][0-9])')
data['posting_year'].describe()
```

Since there is the only posting date year in this given set is 2021 (as shown by .describe()), we are able to simply calculate the age by subtracting the 'year' column from 2021

```python
data['car_age'] = 2021 - (data['year'])
data['car_age'].describe()
```

```python
data.head()
```

Part 4: price_bin column and some of the feature engineering, specifically dummy encoding for our three different types

```python
data['price_bins'] = pd.cut(x=data['price'], bins=[0, 10000, 20000, 25000],
                             labels=['$0-$9,999','$10,000-$19,999','$20,000-$25,000'])
data['price_bins'].describe()
```

```python
data['price_bins'].head(15)
```

Dummy encoding condition, title status, and transmission, and joining them into final dataset

```python
temp = pd.get_dummies(data['condition'], dummy_na=True).astype(int)
temp.head()
```

```python
temp1 = pd.get_dummies(data['title_status'], dummy_na=True).astype(int)
temp1 = temp1.rename(columns={"salvage": "salvage1"})
temp1.head()
```

```python
temp2 = pd.get_dummies(data['transmission'], dummy_na=True).astype(int)
temp2.head()
```

```python
real_df = pd.concat([data, temp, temp1, temp2], axis=1)
real_df.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(
    real_df,
    train_size=0.8,
    random_state=42)

model = LogisticRegression(multi_class='multinomial').fit(
    X=train_data.loc[:, [
        'odometer', 'car_age', 'excellent',
        'fair', 'good', 'like new',
        'new', 'salvage','automatic','manual','other','clean','lien',
        'parts only','rebuilt', 'salvage1']],
    y=train_data['price_bins']
)
```

```python
model.score(
    X=test_data.loc[:, ['odometer', 'car_age', 'excellent',
        'fair', 'good', 'like new',
        'new', 'salvage','automatic','manual','other','clean','lien',
        'parts only','rebuilt', 'salvage1'
        ]],
    y=test_data['price_bins']
)
```
